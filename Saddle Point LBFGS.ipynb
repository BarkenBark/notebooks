```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return x ** 3 - 3 * x * y ** 2

def df(x, y):
    return np.array([
        3 * x ** 2 - 3 * y ** 2,
        - 6 * x * y,
    ])
```

```python
x = np.linspace(-1, 1, 100)
y = np.linspace(-1, 1, 100)

X, Y = np.meshgrid(x, y)
plt.figure(figsize=(10, 10))
plt.contour(X, Y, f(X, Y), 10, cmap=plt.cm.RdBu)
plt.contourf(X, Y, f(X, Y), 100, cmap=plt.cm.RdBu);
```

```python
def eval_params(x):
    return f(x[0], x[1]), df(x[0], x[1])

eval_params([0., 0.])
```

```python
from scipy.optimize import fmin_l_bfgs_b

fmin_l_bfgs_b(eval_params, np.array([1, 0.0001]), disp=10)
```

```python
fmin_l_bfgs_b(eval_params, np.array([-.002, .003]), disp=10)
```

```python
x_small = np.linspace(-.1, .1, 10)
y_small = np.linspace(-.1, .1, 10)

X_small, Y_small = np.meshgrid(x_small, y_small)

rng = np.random.RandomState(42)
pairs = [rng.uniform(-1, 1, size=2) for i in range(500)]

attracted = np.array([
    (x, y) for x, y in pairs
    if fmin_l_bfgs_b(eval_params, np.array([x, y]))[2]['task'].startswith(b'CONVERGENCE: ')
])
attracted
```
