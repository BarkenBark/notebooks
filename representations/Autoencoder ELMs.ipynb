```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_mldata


def load_data(dtype=np.float32, order='F'):
    # Load dataset
    print("Loading dataset...")
    data = fetch_mldata('MNIST original')
    X, y = data.data, data.target
    if order.lower() == 'f':
        X = np.asfortranarray(X)

    # Normalize features to 0-1 range
    X = X.astype('float64')
    X = X / 255

    # Create train-test split (as [Joachims, 2006])
    print("Creating train-test split...")
    n_train = 60000

    X_train = X[:n_train]
    y_train = y[:n_train]
    X_test = X[n_train:]
    y_test = y[n_train:]

    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = load_data(order='C')
```

```python
def plot_images(images, n_rows=10, n_cols=10, ordering=None):
    if ordering is None:
        indices = np.arange(len(images))
    else:
        indices = ordering.argsort()[::-1]
    plt.figure(figsize=(1.2 * n_cols, 1.2 * n_rows))
    plt.gray()
    for i in range(n_rows):
        for j in range(n_cols):
            idx = i * n_cols + j
            if idx < len(images):
                idx = indices[idx]
                plt.subplot(n_rows, n_cols, i * n_cols + j + 1)
                plt.imshow(images[idx].reshape(28, 28), interpolation='nearest')
                plt.axis('off')
                if ordering is not None:
                    plt.title("{:.2f}".format(ordering[idx]))
                
plot_images(X_train)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.linear_model import RidgeCV
from sklearn.utils import check_random_state


class AEELMTransformer(BaseEstimator, TransformerMixin):
    
    def __init__(self, n_components=100, alphas=np.logspace(-3, 3, 7),
                 random_state=None):
        self.n_components = n_components
        self.alphas = alphas
        self.random_state = random_state
        
    def fit(self, X, y=None):
        n_samples, n_features = X.shape
        rng = check_random_state(self.random_state)
        rnd_proj = rng.normal(size=(n_features, self.n_components))
        X_proj = np.tanh(np.dot(X, rnd_proj))
        
        self.ridge_ = ridge = RidgeCV(alphas=self.alphas)
        ridge.fit(X_proj, X)
        
        self.components_ = ridge.coef_.T
        return self
        
    def transform(self, X, y=None):
        return np.tanh(np.dot(X, self.components_.T))
```

```python
from sklearn.cross_validation import train_test_split

n_subsample = 10000
X_orig, _, y_orig, _ = train_test_split(
    X_train, y_train, train_size=n_subsample, random_state=42)
```

```python
%%time

aeelm = AEELMTransformer(n_components=100)
X_transform = aeelm.fit(X_orig).transform(X_orig)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import cross_val_score

def eval_model(clf, X, y, cv=5, n_jobs=1):
    scores = cross_val_score(clf, X, y, cv=cv, n_jobs=n_jobs)
    print("cv score: {:.3f} +/-{:.3f}".format(np.mean(scores), np.std(scores)))
```

```python
%%time
eval_model(LogisticRegression(C=1), X_orig, y_orig, n_jobs=1)
```

```python
%%time
eval_model(LogisticRegression(C=10), X_transform, y_orig, n_jobs=1)
```

```python
%%time
from sklearn.pipeline import Pipeline

p = Pipeline([
    ('aelm1', AEELMTransformer(n_components=100)),
    ('aelm2', AEELMTransformer(n_components=100)),
])
_ = p.fit(X_orig)
```
