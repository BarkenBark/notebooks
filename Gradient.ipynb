```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_0 = 1.2
a_1 = 10.3
a_2 = 2


def f(x, y):
    res_pos = (a_0 * x ** 2) + (a_1 * y ** 2) + (a_2 * x * y)
    res_neg = (a_1 * x ** 2) + (a_0 * y ** 2) + (a_2 * x * y)
    return np.where(x > 0, res_pos, res_neg)


def df(w):
    x, y = w[0], w[1]
    if x > 0:
        return (a_0 * 2 * x) + (a_2 * y), (a_1 * 2 * y) + (a_2 * x)
    else:
        return (a_1 * 2 * x) + (a_2 * y), (a_0 * 2 * y) + (a_2 * x)
    

def gradient_step(w, ctx, step_size=0.1):
    d = df(w)
    update = - step_size * d
    ctx['update'] = update
    ctx['gradient'] = d
    return w + update, np.abs(update).max()


def moment_step(w, ctx, step_size=0.1, moment=0.9):
    update = ctx.get('update', np.zeros_like(w))
    d = df(w)
    update = moment * update - step_size * d
    ctx['update'] = update
    ctx['gradient'] = d
    return w + update, np.abs(update).max()


def nesterov_moment_step(x, y, ctx, step_size=0.1, moment=0.9):
    update_x, update_y = ctx.get('update', (0, 0))
    dx, dy = df(x + moment * update_x, y + moment * update_y)
    update_x = moment * update_x - step_size * dx
    update_y = moment * update_y - step_size * dy
    ctx['update'] = (update_x, update_y)
    ctx['gradient'] = (dx, dy)
    return x + update_x, y + update_y, max(abs(dx), abs(dx))


ada_rate_inc = 1.5
ada_rate_dec = 2

def adasign_step(x, y, ctx, step_size=0.1):
    i = ctx.get('iter', 0)
    step_size_x, step_size_y = ctx.get('step_size', (step_size, step_size))
    update_x, update_y = ctx.get('update', (0, 0))
    prev_dx, prev_dy = ctx.get('gradient', (0, 0))
    dx, dy = df(x, y)
    decrement_x, decrement_y = False, False
    if prev_dx != 0:
        if dx * prev_dx > 0:
            step_size_x *= ada_rate_inc
            step_size_x = min(step_size_x, 1.)
        else:
            step_size_x /= ada_rate_dec
            decrement_x = True
#             print('%03d: step_size_x => %0.3f' % (i, step_size_x))
    
    if prev_dy != 0:
        if dy * prev_dy > 0:
            step_size_y *= ada_rate_inc
            step_size_y = min(step_size_y, 1.)
        else:
            step_size_y /= ada_rate_dec
            decrement_y = True
#             print('%03d: step_size_y => %0.3f' % (i, step_size_y))

    update_x = - step_size_x * dx
#     if decrement_x:
#         update_x /= 2
    update_y = - step_size_y * dy
#     if decrement_y:
#         update_y /= 2
    ctx['position'] = (x, y)
    ctx['gradient'] = (dx, dy)
    ctx['update'] = (update_x, update_y)
    ctx['step_size'] = (step_size_x, step_size_y)
#     print('%03d: (%0.3f, %0.3f) => (%0.3f, %0.3f)' % (i, x, y, x + update_x, y + update_y))
    ctx['iter'] = i + 1
#     print(ctx)
    return x + update_x, y + update_y, max(abs(dx), abs(dy))
```

```python
def plot_convergence(x, y, n_iter=10, color='b', step=gradient_step, tol=1e-3, **kwargs):
    ctx = {}
    for i in range(n_iter):
        new_x, new_y, change = step(x, y, ctx, **kwargs)
        if change < tol:
            print("%s converged in %03d steps" % (step.__name__, i + 1))
            break
        plt.arrow(x, y, new_x - x, new_y - y,
                  width=0.01, length_includes_head=True, alpha=0.2, color=color)
        plt.text(x, y, i, fontsize=12, color=color, clip_on=True,
                 horizontalalignment='center',
                 verticalalignment='center')
        x, y = new_x, new_y
```

```python
x_space = np.linspace(-30, 30, 100)
y_space = np.linspace(-30, 30, 100)

xx, yy = np.meshgrid(x_space, y_space)

plt.figure(figsize=(12, 12))
plt.grid()

extent = [min(x_space), max(x_space), min(y_space), max(y_space)]
objective = f(xx, yy)
plt.imshow(objective, extent=extent, cmap=plt.cm.Reds, origin='lower')
plt.contour(objective, extent=extent, cmap=plt.cm.Reds);

n_iter = 30
step_size = 0.01
# x0, y0 = np.random.uniform(low=-30, high=30, size=2)
x0, y0 = 20, 20


plot_convergence(x0, y0, n_iter=n_iter, color='r', step=gradient_step, step_size=step_size)
plot_convergence(x0, y0, n_iter=n_iter, color='g', step=moment_step, step_size=step_size, moment=0.8)
plot_convergence(x0, y0, n_iter=n_iter, color='b', step=nesterov_moment_step, step_size=step_size, moment=0.8)
plot_convergence(x0, y0, n_iter=n_iter, color='y', step=adasign_step, step_size=step_size)
```

```python
from sklearn.datasets import load_digits
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelBinarizer

digits = load_digits()
X, y = shuffle(digits.data, digits.target, random_state=0)
X = MinMaxScaler().fit_transform(X)
Y = LabelBinarizer(neg_label=-1, pos_label=1).fit_transform(y)


class NN(object):
    
    def __init__(self, n_nodes=(64, 100, 100, 100, 10), batch_size=100, seed=1):
        rng = np.random.RandomState(seed)
        self.weights = []
        self.intercepts = []
        self.activations = []
        for i in range(len(n_nodes) - 1):
            self.weights.append(rng.randn(n_nodes[i], n_nodes[i + 1]))
            self.intercepts.append(rng.randn(n_nodes[i + 1]) + 0.01)
            self.activations.append(np.empty(shape=(batch_size, n_nodes[i + 1]), dtype=np.float64))
    
    def forward(self, data):
        for i, activation in enumerate(self.activations):
            data = np.dot(data, self.weights[i])
            data += self.intercepts[i]
            np.clip(data, 0, np.finfo(data.dtype).max, out=data)  # inplace relu
            activation[:] = data
        return data
    
    def loss(self, data, target):
        """OvA squared hinge loss"""
        predicted = self.forward(data)
        # target is expected to be binarized in range {-1, 1}
        return np.clip(1 - predicted * target, 0, np.finfo(data.dtype).max) ** 2
        
nn = NN(batch_size=3)
nn.forward(X[:3])
```

```python
def loss(nn, data, target, w_x, w_y, x_idx=(1, 0, 0), y_idx=(3, 1, 1)):
    results = []
    for w_x_i, w_y_i in zip(w_x.ravel(), w_y.ravel()):
        nn.weights[x_idx[0]][x_idx[1:2]] = w_x_i
        nn.weights[y_idx[0]][y_idx[1:2]] = w_y_i
        results.append(nn.loss(data, target).sum())
    return np.array(results).reshape(w_x.shape)
```

```python
from sklearn.cross_validation import train_test_split

batch_size = 30
nn = NN(batch_size=batch_size)

radius = 50
resolution = 20
w_x_space = np.linspace(-radius, radius, resolution)
w_y_space = np.linspace(-radius, radius, resolution)

w_xx, w_yy = np.meshgrid(w_x_space, w_y_space)
extent = [min(w_x_space), max(w_x_space), min(w_y_space), max(w_y_space)]

plt.figure(figsize=(16, 4))
n_plots = 8
for i in range(n_plots):
    plt.subplot(1, n_plots, i + 1)
    X_batch, _, Y_batch, _ = train_test_split(X, Y, train_size=batch_size,
                                              random_state=i)
    plt.grid()
    objective = loss(nn, X_batch, Y_batch, w_xx, w_yy)
    plt.imshow(objective, extent=extent, cmap=plt.cm.Reds, origin='lower')
    plt.contour(objective, extent=extent, cmap=plt.cm.Reds)
    ax = plt.gca()
    ax.xaxis.set_major_formatter(plt.NullFormatter())
    ax.yaxis.set_major_formatter(plt.NullFormatter())
```
