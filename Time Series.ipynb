```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.RandomState(0)
```

```python
n_steps = 1000
time = np.arange(n_steps)

data_deterministic = (
    np.sin(time / 10.)
    + 0.7 * np.sin(time / 12. + 5) * np.sin(time / 5.))

data_noisy = (
    data_deterministic
    + 0.1 * np.cumsum(rng.normal(size=n_steps)))

plt.plot(data_deterministic)
plt.plot(data_noisy)
plt.ylim(-4, 4)
```

```python
target = data_deterministic[5:]

features = np.hstack([
    np.diff(data_deterministic[3:-1]).reshape(-1, 1),
    np.max(data_deterministic[3:-1])
    data_noisy[4:-1].reshape(-1, 1),
    data_noisy[3:-2].reshape(-1, 1),
    data_noisy[2:-3].reshape(-1, 1),
    data_noisy[1:-4].reshape(-1, 1),
    data_noisy[0:-5].reshape(-1, 1),
])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import cross_val_score
```

```python
scores = cross_val_score(LinearRegression(), features, target, cv=5,
                         scoring='mean_squared_error')
np.mean(scores), np.std(scores)
```

```python
from sklearn.ensemble import ExtraTreesRegressor

scores = cross_val_score(ExtraTreesRegressor(n_estimators=10), features, target, cv=5,
                         scoring='mean_squared_error')
np.mean(scores), np.std(scores)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

scores = cross_val_score(GradientBoostingRegressor(n_estimators=5, max_depth=3), features, target, cv=5,
                         scoring='mean_squared_error')
np.mean(scores), np.std(scores)
```

```python
from sklearn.svm import SVR

scores = cross_val_score(SVR(gamma=0.01, C=.1), features, target, cv=5,
                         scoring='mean_squared_error')
np.mean(scores), np.std(scores)
```
