```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_data = np.array([
    [1., -1],
    [2.5, -2],
    [4., -3],
    [-1.5, 1],
    [-1.5, 2],
    [-2.0, 3],
]).astype(np.float64)
y_data = np.array([
    1,
    1,
    1,
    -1,
    -1,
    -1,
]).astype(np.float64)


def f(all_x, all_y):
    results = []
    for x, y in zip(all_x.ravel(), all_y.ravel()):
        w = np.array([x, y]).reshape(2, 1)
        hinge_loss = np.sum(np.clip(-np.dot(X_data, w) * y + 1, 0, 1e24))
        results.append(hinge_loss)
    return np.array(results).reshape(all_x.shape)


def df(x, y):
    w = np.array([x, y]).reshape(2, 1)
    if np.sum(np.clip(-np.dot(X_data, w) * y + 1, 0, 1e24)) > 0:
        return - X_data.sum(axis=0)
    else:
        return np.zeros(2)

a_0 = 1.2
a_1 = 10.3
a_2 = 2


def f(x, y):
    res_pos = (a_0 * x ** 2) + (a_1 * y ** 2) + (a_2 * x * y)
    res_neg = (a_1 * x ** 2) + (a_0 * y ** 2) + (a_2 * x * y)
    return np.where(x > 0, res_pos, res_neg)


def df(x, y):
    if x > 0:
        return (a_0 * 2 * x) + (a_2 * y), (a_1 * 2 * y) + (a_2 * x)
    else:
        return (a_1 * 2 * x) + (a_2 * y), (a_0 * 2 * y) + (a_2 * x)
    

def gradient_step(x, y, ctx, step_size=0.1):
    dx, dy = df(x, y)
    update_x = - step_size * dx
    update_y = - step_size * dy
    ctx['gradient'] = (dx, dy)
    return x + update_x, y + update_y, max(abs(update_x), abs(update_y))


def moment_step(x, y, ctx, step_size=0.1, moment=0.9):
    update_x, update_y = ctx.get('update', (0, 0))
    dx, dy = df(x, y)
    update_x = moment * update_x - step_size * dx
    update_y = moment * update_y - step_size * dy
    ctx['update'] = (update_x, update_y)
    ctx['gradient'] = (dx, dy)
    return x + update_x, y + update_y, max(abs(update_x), abs(update_y))


def nesterov_moment_step(x, y, ctx, step_size=0.1, moment=0.9):
    update_x, update_y = ctx.get('update', (0, 0))
    dx, dy = df(x + moment * update_x, y + moment * update_y)
    update_x = moment * update_x - step_size * dx
    update_y = moment * update_y - step_size * dy
    ctx['update'] = (update_x, update_y)
    ctx['gradient'] = (dx, dy)
    return x + update_x, y + update_y, max(abs(update_x), abs(update_y))


ada_rate_inc = 1.5
ada_rate_dec = 2

def adasign_step(x, y, ctx, step_size=0.1):
    i = ctx.get('iter', 0)
    step_size_x, step_size_y = ctx.get('step_size', (step_size, step_size))
    update_x, update_y = ctx.get('update', (0, 0))
    prev_dx, prev_dy = ctx.get('gradient', (0, 0))
    dx, dy = df(x, y)
    if prev_dx != 0:
        if dx * prev_dx > 0:
            step_size_x *= ada_rate_inc
            step_size_x = min(step_size_x, 1.)
        else:
            step_size_x /= ada_rate_dec
#             print('%03d: step_size_x => %0.3f' % (i, step_size_x))
    
    if prev_dy != 0:
        if dy * prev_dy > 0:
            step_size_y *= ada_rate_inc
            step_size_y = min(step_size_y, 1.)
        else:
            step_size_y /= ada_rate_dec
#             print('%03d: step_size_y => %0.3f' % (i, step_size_y))

    update_x = - step_size_x * dx
    update_y = - step_size_y * dy
    ctx['position'] = (x, y)
    ctx['gradient'] = (dx, dy)
    ctx['update'] = (update_x, update_y)
    ctx['step_size'] = (step_size_x, step_size_y)
#     print('%03d: (%0.3f, %0.3f) => (%0.3f, %0.3f)' % (i, x, y, x + update_x, y + update_y))
    ctx['iter'] = i + 1
#     print(ctx)
    return x + update_x, y + update_y, max(abs(update_x), abs(update_y))


def adasign2_step(x, y, ctx, step_size=0.1):
    i = ctx.get('iter', 0)
    step_size_x, step_size_y = ctx.get('step_size', (step_size, step_size))
    update_x, update_y = ctx.get('update', (0, 0))
    prev_dx, prev_dy = ctx.get('gradient', (0, 0))
    prev_x, prev_y = ctx.get('position', (0, 0))
    dx, dy = df(x, y)
#     print('%03d: pos: (%0.3f, %0.3f), grad: (%0.3f, %0.3f)' % (i, x, y, dx, dy))
    backtrack = False
    if prev_dx != 0:
        if dx * prev_dx < 0:
            step_size_x /= ada_rate_dec
            backtrack = True
#             print('%03d: step_size_x => %0.3f' % (i, step_size_x))
    if prev_dy != 0:
        if dy * prev_dy < 0:
            step_size_y /= ada_rate_dec
            backtrack = True
#             print('%03d: step_size_y => %0.3f' % (i, step_size_y))
    if not backtrack and (prev_dx != 0 or prev_dy != 0):
        # all the steps
        step_size_x *= ada_rate_inc
        step_size_x = min(step_size_x, 1.)
        step_size_y *= ada_rate_inc
        step_size_y = min(step_size_y, 1.)

    if backtrack:
        x, y = prev_x, prev_y
        dx, dy = prev_dx, prev_dy

    update_x = - step_size_x * dx
    update_y = - step_size_y * dy
    ctx['gradient'] = (dx, dy)
    ctx['position'] = (x, y)
    ctx['update'] = (update_x, update_y)
    ctx['step_size'] = (step_size_x, step_size_y)
#     print('%03d: (%0.3f, %0.3f) => (%0.3f, %0.3f)' % (i, x, y, x + update_x, y + update_y))
    ctx['iter'] = i + 1
#     print(ctx)
    return x + update_x, y + update_y, max(abs(update_x), abs(update_y))
```

```python
def plot_convergence(x, y, n_iter=10, color='b', step=gradient_step, tol=1e-3, **kwargs):
    ctx = {}
    for i in range(n_iter):
        new_x, new_y, change = step(x, y, ctx, **kwargs)
        if change < tol:
            print("%s converged in %03d steps" % (step.__name__, i + 1))
            break
        plt.arrow(x, y, new_x - x, new_y - y,
                  width=0.005, length_includes_head=True, alpha=0.5, color=color)
#         plt.text(x, y, i, fontsize=12, color=color)
        x, y = new_x, new_y
```

```python
plt.figure(figsize=(12, 12))
x_space = np.linspace(-30, 30, 100)
y_space = np.linspace(-30, 30, 100)

xx, yy = np.meshgrid(x_space, y_space)
plt.grid()
plt.contour(x_space, y_space, f(xx, yy))

n_iter = 100
step_size = 0.001
x0, y0 = np.random.uniform(low=-30, high=30, size=2)
# x0, y0 = 20, 20
plot_convergence(x0, y0, n_iter=n_iter, color='r', step=gradient_step, step_size=step_size)
plot_convergence(x0, y0, n_iter=n_iter, color='g', step=moment_step, step_size=step_size, moment=0.8)
plot_convergence(x0, y0, n_iter=n_iter, color='b', step=nesterov_moment_step, step_size=step_size, moment=0.8)
plot_convergence(x0, y0, n_iter=n_iter, color='y', step=adasign_step, step_size=step_size)
```
